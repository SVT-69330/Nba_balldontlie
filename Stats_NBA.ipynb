# NBA Statistique

## Import Module

```python
import requests
import pandas as pd
from google.colab import userdata
```

## Variable


```python
# variable Google
API_NBA = userdata.get('API_ballondontlie')

# variable Api balldontlie
headers = {"Authorization": API_NBA}
url_Teams = "https://api.balldontlie.io/v1/teams"
url_Games = "https://api.balldontlie.io/v1/games"
url_Players = "https://api.balldontlie.io/v1/players"
url_GamesPlayerStats = "https://api.balldontlie.io/v1/stats"

```

## Connexions

### Connexion Teams

```python
# Connexion au Endpoint
r_Teams = requests.get(url_Teams, headers=headers)
# r = requests.get(url, params=params, headers=headers)
data_Teams = r_Teams.json()
#print(data)
pd_data_Teams = pd.DataFrame(data_Teams['data'])
pd_data_Teams.head()
#
```

### Connexion Players


```python

# Connexion au Endpoint
r_Players  = requests.get(url_Players , headers=headers)
# r = requests.get(url, params=params, headers=headers)
data_Players  = r_Players.json()
#print(data)
pd_data_Players = pd.DataFrame(data_Players ['data'])
pd_data_Players.head()
```

### Connexion Games

```python
# Connexion au Endpoint
r_Games  = requests.get(url_Games , headers=headers)
# print(r_Games)
# r = requests.get(url, params=params, headers=headers)
data_Games  = r_Games.json()
#print(data)
pd_data_Games  = pd.DataFrame(data_Games ['data'])
pd_data_Games.head()
```

## Prparation Datas

### Connexion DF Teams et Players

```python
pd_data_Players['team_id'] = pd_data_Players['team'].apply(lambda x: x['id'] if isinstance(x, dict) else None)
merged_df = pd.merge(pd_data_Players, pd_data_Teams, left_on='team_id', right_on='id', suffixes=('_player', '_team'))
display(merged_df.head())
```

```python
columns_to_drop = ['team', 'team_id']
pd_data_teams_py = merged_df.drop(columns=columns_to_drop, errors='ignore')
display(pd_data_teams_py.head())
```
