## Connexions

### Connexion Teams

```python


# API_BASE_URL = 'https://www.balldontlie.io/api/v1/teams'

url_teams = "https://api.balldontlie.io/v1/teams"
# params = {"full_name": "Phoenix Suns"}
headers = {"Authorization": 'c3963ba9-06fb-4a99-97cc-580664807ae6'}

r_teams = requests.get(url_teams, headers=headers)
# r = requests.get(url, params=params, headers=headers)
data_teams = r_teams.json()
#print(data)
pd_data_teams = pd.DataFrame(data_teams['data'])
pd_data_teams.head()
#
```

### Connexion Players


```python

url_py = "https://api.balldontlie.io/v1/players"
# params = {"full_name": "Phoenix Suns"}

r_py  = requests.get(url_py , headers=headers)
# r = requests.get(url, params=params, headers=headers)
data_py  = r_py .json()
#print(data)
pd_data_py  = pd.DataFrame(data_py ['data'])
pd_data_py.head()
```

### Connexion Games

```python
url_Games = "https://api.balldontlie.io/v1/games"
# params = {"full_name": "Phoenix Suns"}

r_Games  = requests.get(url_Games , headers=headers)
# r = requests.get(url, params=params, headers=headers)
data_Games  = r_Games .json()
#print(data)
pd_data_Games  = pd.DataFrame(data_Games ['data'])
pd_data_Games.head()
```
